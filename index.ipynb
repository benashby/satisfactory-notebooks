## How this works

Each notebook is a single blueprint step. **Producers** push into a dimensional depot/storage container. **Consumers** pull from that storage — always pulling *less* than what's being pushed, so the storage stays full.

The gap between production and consumption is your **stash rate** — a buffer that keeps you safe if a line hiccups.

## Ingot budget

Fill this in as you set up each notebook. Total must not exceed 270.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path('.').resolve()))

import pandas as pd

# ── Ingot allocations — fill in as each notebook is completed ─────────────────
allocations = [
    ('Iron Rods',               None),  # ← from iron-rods.ipynb: IRON_INGOTS
    ('Iron Plates',             None),  # ← from iron-plates.ipynb: IRON_INGOTS
    ('Screws',                  None),  # ← from screws.ipynb: IRON_INGOTS
]

TOTAL_SUPPLY = 270.0

filled = [(name, rate) for name, rate in allocations if rate is not None]
total_allocated = sum(r for _, r in filled)
remaining = TOTAL_SUPPLY - total_allocated

rows = [(name, f'{rate:.1f}' if rate is not None else '—') for name, rate in allocations]
rows.append(('──────────────', '──────'))
rows.append(('Total allocated', f'{total_allocated:.1f}'))
rows.append(('Remaining',       f'{remaining:.1f}'))
rows.append(('Supply',          f'{TOTAL_SUPPLY:.1f}'))

df = pd.DataFrame(rows, columns=['Chain', 'Iron Ingots/min'])
df = df.set_index('Chain')
df
```
