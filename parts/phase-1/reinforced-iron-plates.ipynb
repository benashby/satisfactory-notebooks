# Reinforced Iron Plates

**Type:** Consumer — pulls from storage, stores output in dimensional depot  
**Blueprint:** `reinforced-iron-plate` (90 Iron Plates/min + 180 Screws/min → 15 Reinforced Iron Plates/min at 100%)  
**Pulls from:** Iron Plates storage, Screws storage  
**Downstream:** Smart Plating pulls from RIP storage (stashing 5 Reinforced Iron Plates/min)

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path('../..').resolve()))

from blueprints import BLUEPRINTS, Blueprint, Stage
from satisfactory import ITEMS

BP  = BLUEPRINTS
TOL = 0.5
```

## Constants

```python
# ── Supply from storage ───────────────────────────────────────────────────────
# iron-plates produces 212.91/min, stashes 20/min → 192.91 available
IRON_PLATES_FROM_STORAGE = 192.91  # Iron Plates/min pulled from storage
# screws produces 1239.60/min; Rotors take 804.00/min → 385.60 available for RIPs
SCREWS_FROM_STORAGE      = 385.92  # Screws/min pulled from storage

# ── Stash target ──────────────────────────────────────────────────────────────
STASH_RATE = 5.0   # Reinforced Iron Plates/min accumulating in storage

# ── Blueprint placement ───────────────────────────────────────────────────────
REINFORCED_PLATE_COPIES      = 3
REINFORCED_PLATE_OUTPUT_EACH = 10.72  # Reinforced Iron Plates/min — total per blueprint copy
```

## Derived rates and balance

```python
reinforced_total = REINFORCED_PLATE_COPIES * REINFORCED_PLATE_OUTPUT_EACH
plates_consumed  = reinforced_total * BP['reinforced-iron-plate'].input_ratio('iron-plate', 'reinforced-iron-plate')
screws_consumed  = reinforced_total * BP['reinforced-iron-plate'].input_ratio('screw',      'reinforced-iron-plate')

available_for_smart_plating = reinforced_total - STASH_RATE

assert abs(plates_consumed - IRON_PLATES_FROM_STORAGE) < TOL, (
    f"Iron Plate balance: consuming {plates_consumed:.2f}/min, available {IRON_PLATES_FROM_STORAGE:.2f}/min "
    f"(delta {plates_consumed - IRON_PLATES_FROM_STORAGE:+.2f})"
)
assert abs(screws_consumed - SCREWS_FROM_STORAGE) < TOL, (
    f"Screw balance: consuming {screws_consumed:.2f}/min, available {SCREWS_FROM_STORAGE:.2f}/min "
    f"(delta {screws_consumed - SCREWS_FROM_STORAGE:+.2f})"
)
assert available_for_smart_plating > 0

print(f"✓ Chain balanced")
print(f"  {IRON_PLATES_FROM_STORAGE:.2f} Iron Plates/min  +  {SCREWS_FROM_STORAGE:.2f} Screws/min")
print(f"  →  {reinforced_total:.2f} Reinforced Iron Plates/min ({REINFORCED_PLATE_COPIES} blueprint copies × {REINFORCED_PLATE_OUTPUT_EACH:.2f}/min each)")
print(f"  Stashing:                    {STASH_RATE:.0f} Reinforced Iron Plates/min")
print(f"  Available for Smart Plating: {available_for_smart_plating:.2f} Reinforced Iron Plates/min")
print()
print(f"  ⚠ Machine counts not yet registered for this blueprint.")
print(f"  Set each assembler inside to: {REINFORCED_PLATE_OUTPUT_EACH:.2f} ÷ [assemblers in blueprint] Reinforced Iron Plates/min")
```
